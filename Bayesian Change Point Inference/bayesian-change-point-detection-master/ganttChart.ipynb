```python
# Update the tasks with new durations for Stage 2, Stage 3, and Stage 4
tasks = [
    {"name": "Stage 1: Brute-force with single distribution", "start": datetime(2025, 1, 20), "duration": 14, "color": "lightgreen"},
    {"name": "Stage 2: Overlapping distributions", "start": datetime(2025, 2, 3), "duration": 14, "color": "yellow"},
    {"name": "Stage 3: Two distinct distributions", "start": datetime(2025, 2, 17), "duration": 14, "color": "lightcoral"},
    {"name": "Stage 4: Searching method", "start": datetime(2025, 3, 3), "duration": 30, "color": "lightskyblue"},
]

# Calculate task end dates
for task in tasks:
    task["end"] = task["start"] + timedelta(days=task["duration"])
weeks = [datetime(2025, 1, 20) + timedelta(weeks=i) for i in range(12)]  # Extend timeline to cover longer duration

# Create the figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Add task bars
for i, task in enumerate(tasks):
    ax.barh(
        y=task["name"],
        left=date2num(task["start"]),
        width=date2num(task["end"]) - date2num(task["start"]),
        color=task["color"],
        edgecolor="black",
        height=0.4,
    )

# Add week labels at the top
ax.set_xticks([date2num(week) for week in weeks])
ax.set_xticklabels([f"Week {i+1}\n{week.strftime('%d %b')}" for i, week in enumerate(weeks)], fontsize=10)
ax.tick_params(axis="x", length=0)  # Remove tick marks
ax.invert_yaxis()  # Align tasks from top to bottom

# Add gridlines for clarity
ax.xaxis.grid(True, linestyle="--", alpha=0.5)
ax.set_xlim(date2num(datetime(2025, 1, 20)), date2num(datetime(2025, 4, 20)))  # Adjust x-axis limits

# Add chart title and labels
ax.set_title("Gantt Chart for Project Stages", fontsize=14, weight="bold")
ax.set_xlabel("Timeline", fontsize=12)
ax.set_ylabel("Stages", fontsize=12)

# Adjust layout for better readability
plt.tight_layout()

# Display the Gantt chart
plt.show()
```
