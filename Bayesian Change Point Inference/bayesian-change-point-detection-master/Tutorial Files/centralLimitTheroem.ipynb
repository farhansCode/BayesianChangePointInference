```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 100000  # Size of the population
sample_sizes = [5, 30, 100]  # Sample sizes to test
num_samples = 1000  # Number of samples to draw for each sample size

# Generate a population of dice rolls (values 1 to 6)
population = np.random.randint(1, 7, size=population_size)

# Plot the population distribution
plt.figure(figsize=(10, 6))
plt.hist(population, bins=np.arange(1, 8) - 0.5, color='skyblue', alpha=0.7, label='Population')
plt.title('Population Distribution (Dice Rolls)')
plt.xlabel('Dice Value')
plt.ylabel('Frequency')
plt.xticks(range(1, 7))
plt.legend()
plt.show()

# Observe the CLT with different sample sizes
for n in sample_sizes:
    sample_means = []

    # Draw samples and calculate their means
    for _ in range(num_samples):
        sample = np.random.choice(population, size=n, replace=True)
        sample_means.append(np.mean(sample))

    # Plot the distribution of sample means
    plt.figure(figsize=(10, 6))
    plt.hist(sample_means, bins=30, color='orange', alpha=0.7, label=f'Sample Size = {n}')
    plt.title(f'Distribution of Sample Means (Sample Size = {n})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```
